```python
import pandas as pd
from google.colab import files


uploaded = files.upload()


file_name = list(uploaded.keys())[0]
df = pd.read_csv(file_name)


df.head()
```

```python

df.info()


missing_values = df.isnull().sum()
print("Missing Values:\n", missing_values)


df_cleaned = df.dropna(subset=["Total"])


df_cleaned.info()
```

```python

duplicates = df_cleaned.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")


df_cleaned = df_cleaned.drop_duplicates()
```

```python

df_cleaned["Date"] = pd.to_datetime(df_cleaned["Date"], errors="coerce")


print(df_cleaned["Date"].head())


df_cleaned["Time"] = pd.to_datetime(df_cleaned["Time"], format='%I:%M %p', errors='coerce').dt.time


df_cleaned.head()
```

```python

df_cleaned.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


branch_counts = df_cleaned["Branch"].value_counts()


sns.barplot(x=branch_counts.index, y=branch_counts.values, palette="mako")
plt.title("Distribution of Sales by Branch")
plt.xlabel("Branch")
plt.ylabel("Number of Sales")
plt.show()
```

```python

payment_counts = df_cleaned["Payment"].value_counts()


sns.barplot(x=payment_counts.index, y=payment_counts.values, palette="coolwarm")
plt.title("Payment Methods Distribution")
plt.xlabel("Payment Method")
plt.ylabel("Number of Transactions")
plt.show()
```

```python

plt.figure(figsize=(8, 5))
sns.scatterplot(x="Quantity", y="Unit price", data=df_cleaned, hue="Branch", palette="viridis")
plt.title("Relationship between Quantity and Unit Price")
plt.xlabel("Quantity")
plt.ylabel("Unit Price")
plt.show()
```

```python

sns.histplot(df_cleaned["Rating"], bins=10, kde=True, color="purple")
plt.title("Customer Ratings Distribution")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.show()
```

```python
from fpdf import FPDF

class PDF(FPDF):
    def header(self):
        self.set_font('Arial', 'B', 12)
        self.cell(0, 10, 'Data Wrangling Report', align='C', ln=1)
        self.ln(5)

    def chapter_title(self, title):
        self.set_font('Arial', 'B', 12)
        self.cell(0, 10, title, ln=1)
        self.ln(5)

    def chapter_body(self, body):
        self.set_font('Arial', '', 12)
        self.multi_cell(0, 10, body)
        self.ln()


pdf = PDF()
pdf.add_page()


pdf.chapter_title("1. Summary of Data Cleaning")


cleaning_summary = f"""
The dataset originally had {df.shape[0]} rows and {df.shape[1]} columns.
After cleaning:
- Missing values were removed.
- Duplicate rows were dropped.
- Date and time formats were standardized.

Final dataset:
- {df_cleaned.shape[0]} rows
- {df_cleaned.shape[1]} columns

Key attributes cleaned:
- 'Total' column with missing values
- Duplicates removed from the dataset
"""
pdf.chapter_body(cleaning_summary)


pdf.output("Data_Wrangling_Report.pdf")
```

```python

pdf = PDF()
pdf.add_page()


pdf.chapter_title("2. Business Insights")

insights_summary = f"""
1. Branch Analysis:
   - Most sales are in branch '{branch_counts.idxmax()}' with {branch_counts.max()} transactions.

2. Payment Methods:
   - The most used payment method is '{payment_counts.idxmax()}' with {payment_counts.max()} transactions.

3. Quantity vs Unit Price:
   - A higher quantity of items tends to have diverse unit prices, as shown in the scatter plot.

4. Customer Ratings:
   - The average customer rating is {df_cleaned['Rating'].mean():.2f}, indicating a positive customer experience overall.
"""
pdf.chapter_body(insights_summary)


pdf.output("Business_Insights_Report.pdf")
```
